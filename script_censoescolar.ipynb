```python
# Carregando e entendendo o objeto que continha os municípios que precisarei filtrar
municipios = pd.read_csv('totalGruposMunicipios.csv')
municipios = municipios['municipio']
print(len(municipios.unique()))
print(municipios)
```

```python
dicio = {} #Dicionário vazio para armazenar as bases prontas ao final
agregacao = {} #Dicionário que conterá as informações para agregação por municípios das bases( Está por escola)

# Leitura dos dados de cada arquivo
for arquivo in os.listdir():
    if arquivo.startswith(('base')):
        caminho_dados = os.path.basename(arquivo)
        data = (os.path.basename(arquivo))[-8:-4]
        nome_arquivo = f"censo_{data}"
        dados = pd.read_csv(caminho_dados, sep=',', encoding='latin-1')
        dados = dados[dados['NO_MUNICIPIO'].isin(municipios)]
        dados = dados.rename(columns={"Ãndice_de_Infraestrutura_Escolar": 'Indice_infra'})
        dados = dados.drop(columns = ['NO_ENTIDADE']).copy()
        dados['ano'] = data
        for column in dados.columns:
            if column.startswith('QT'):
                agregacao[column] = (column,'sum')
            else: 
                agregacao[column] = (column,'unique')
            agregacao['Indice_infra'] =('Indice_infra','mean')
        dados = dados.groupby("NO_MUNICIPIO").agg(**agregacao)
        dicio[nome_arquivo] = dados
        # Check de adição ao dicionário
        print(f"Arquivo '{nome_arquivo}' adicionado ao dicionário!")
```
